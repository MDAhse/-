```python
df2 = df.dropna(subset=["year"])
df2.loc[:, "year"] = df2["year"].apply(lambda x: x)
df2.loc[:, "artist_country"] = df2["artist_country"].apply(lambda x: x)
df2.loc[:, "rank"] = df2["rank"].apply(lambda x: x)
```

```python
df2 = df2.pivot_table(index="artist_country", columns="year", values="rank")
df2
```

```python
df3 = df2.div(df2.sum(axis=1), axis=0)
df3
```
